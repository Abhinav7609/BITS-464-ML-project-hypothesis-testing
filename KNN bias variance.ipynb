```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from mlxtend.evaluate import bias_variance_decomp
```

Synthetic Dataset for KNN

```python

def dataset_KNN(num_samples):
 
    X1 = np.random.uniform(low=-1, high=1, size=num_samples).reshape(-1, 1)
    X2 = np.random.uniform(low=-1, high=1, size=num_samples).reshape(-1, 1)
    
   
    X = np.concatenate((X1, X2), axis=1)
    
   
    y = np.exp(X1) +2*(X2) + np.random.normal(scale=0.5, size=num_samples).reshape(-1, 1)
    
    return X, y

# Generate synthetic data
X_train_knn, y_train_knn = dataset_KNN(1000)
X_test_knn,y_test= dataset_KNN(1000)
```

```python
H0=KNeighborsRegressor(n_neighbors=10)
H1=KNeighborsRegressor(n_neighbors=5)
H2=KNeighborsRegressor(n_neighbors=3)
H3=KNeighborsRegressor(n_neighbors=2)
H4=KNeighborsRegressor(n_neighbors=1)
H_neg1=KNeighborsRegressor(n_neighbors=15)
H_neg2=KNeighborsRegressor(n_neighbors=20)
H_neg3=KNeighborsRegressor(n_neighbors=25)
H_neg4=KNeighborsRegressor(n_neighbors=50)
H_neg5=KNeighborsRegressor(n_neighbors=100)

H0.fit(X_train_knn, y_train_knn)
H1.fit(X_train_knn, y_train_knn)
H2.fit(X_train_knn, y_train_knn)
H3.fit(X_train_knn, y_train_knn)
H4.fit(X_train_knn, y_train_knn)
H_neg1.fit(X_train_knn, y_train_knn)
H_neg2.fit(X_train_knn, y_train_knn)
H_neg3.fit(X_train_knn, y_train_knn)
H_neg4.fit(X_train_knn, y_train_knn)
H_neg5.fit(X_train_knn, y_train_knn)
```

```python
H0_pred=H0.predict(X_test_knn)
H1_pred = H1.predict(X_test_knn)
H2_pred = H2.predict(X_test_knn)
H3_pred = H3.predict(X_test_knn)
H4_pred = H4.predict(X_test_knn)
H_neg1_pred = H_neg1.predict(X_test_knn)
H_neg2_pred = H_neg2.predict(X_test_knn)
H_neg3_pred = H_neg3.predict(X_test_knn)
H_neg4_pred = H_neg4.predict(X_test_knn)
H_neg5_pred = H_neg5.predict(X_test_knn)
```

```python
mse_h0 = mean_squared_error(y_test, H0_pred)
mse_h1 = mean_squared_error(y_test, H1_pred)
mse_h2 = mean_squared_error(y_test, H2_pred)
mse_h3 = mean_squared_error(y_test, H3_pred)
mse_h4 = mean_squared_error(y_test, H4_pred)
mse_h_neg1 = mean_squared_error(y_test, H_neg1_pred)
mse_h_neg2 = mean_squared_error(y_test, H_neg2_pred)
mse_h_neg3 = mean_squared_error(y_test, H_neg3_pred)
mse_h_neg4 = mean_squared_error(y_test, H_neg4_pred)
mse_h_neg5 = mean_squared_error(y_test, H_neg5_pred)
```

```python
print("MSE for H0:", mse_h0)
print("MSE for H1:", mse_h1)
print("MSE for H2:", mse_h2)
print("MSE for H3:", mse_h3)
print("MSE for H4:", mse_h4)
print("MSE for H_neg1:", mse_h_neg1)
print("MSE for H_neg2:", mse_h_neg2)
print("MSE for H_neg3:", mse_h_neg3)
print("MSE for H_neg4:", mse_h_neg4)
print("MSE for H_neg5:", mse_h_neg5)
```

```python
train_mse_h0 = mean_squared_error(y_train_knn, H0.predict(X_train_knn))
train_mse_h1 = mean_squared_error(y_train_knn, H1.predict(X_train_knn))
train_mse_h2 = mean_squared_error(y_train_knn, H2.predict(X_train_knn))
train_mse_h3 = mean_squared_error(y_train_knn, H3.predict(X_train_knn))
train_mse_h4 = mean_squared_error(y_train_knn, H4.predict(X_train_knn))
train_mse_h_neg1 = mean_squared_error(y_train_knn, H_neg1.predict(X_train_knn))
train_mse_h_neg2 = mean_squared_error(y_train_knn, H_neg2.predict(X_train_knn))
train_mse_h_neg3 = mean_squared_error(y_train_knn, H_neg3.predict(X_train_knn))
train_mse_h_neg4 = mean_squared_error(y_train_knn, H_neg4.predict(X_train_knn))
train_mse_h_neg5 = mean_squared_error(y_train_knn, H_neg5.predict(X_train_knn))


print("Training MSE for H0:", train_mse_h0)
print("Training MSE for H1:", train_mse_h1)
print("Training MSE for H2:", train_mse_h2)
print("Training MSE for H3:", train_mse_h3)
print("Training MSE for H4:", train_mse_h4)
print("Training MSE for H_neg1:", train_mse_h_neg1)
print("Training MSE for H_neg2:", train_mse_h_neg2)
print("Training MSE for H_neg3:", train_mse_h_neg3)
print("Training MSE for H_neg4:", train_mse_h_neg4)
print("Training MSE for H_neg5:", train_mse_h_neg5)
```

```python



k_values = [1, 2, 3, 5, 10, 15, 20, 25, 50, 100] 
train_errors = [train_mse_h4, train_mse_h3, train_mse_h2, train_mse_h1, train_mse_h0, train_mse_h_neg1, train_mse_h_neg2, train_mse_h_neg3, train_mse_h_neg4, train_mse_h_neg5] 
test_errors = [mse_h4, mse_h3, mse_h2, mse_h1, mse_h0, mse_h_neg1, mse_h_neg2, mse_h_neg3, mse_h_neg4, mse_h_neg5]  


plt.figure(figsize=(8, 6))
plt.plot(k_values, train_errors, marker='o', linestyle='-', color='r', label='Train MSE')
plt.plot(k_values, test_errors, marker='o', linestyle='-', color='b', label='Test MSE')
plt.xlabel('Value of k')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Train and Test Error Rate vs. k Value for KNN Regressor Models')
plt.grid(True)
plt.xticks(k_values)  
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

original_k_values = [1, 2, 3, 5, 10, 15, 20, 25, 50, 100]
train_errors = [train_mse_h4, train_mse_h3, train_mse_h2, train_mse_h1, train_mse_h0, train_mse_h_neg1, train_mse_h_neg2, train_mse_h_neg3, train_mse_h_neg4, train_mse_h_neg5]
test_errors = [mse_h4, mse_h3, mse_h2, mse_h1, mse_h0, mse_h_neg1, mse_h_neg2, mse_h_neg3, mse_h_neg4, mse_h_neg5]

reversed_k_values = original_k_values[::-1]


one_over_k_values = [1 / k for k in original_k_values]

plt.figure(figsize=(10, 6))
plt.plot(reversed_k_values, train_errors, marker='o', linestyle='-', color='r', label='Train MSE')
plt.plot(reversed_k_values, test_errors, marker='o', linestyle='-', color='b', label='Test MSE')
plt.xlabel('Reversed k values')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Train and Test Error Rate vs. Reversed k Value for KNN Regressor Models')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(one_over_k_values, train_errors, marker='o', linestyle='-', color='r', label='Train MSE')
plt.plot(one_over_k_values, test_errors, marker='o', linestyle='-', color='b', label='Test MSE')
plt.xlabel('1/k values')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Train and Test Error Rate vs. 1/k Value for KNN Regressor Models')
plt.grid(True)
plt.legend()
plt.show()
```

```python

y_train_knn = y_train_knn.flatten()
models = [
    H_neg3,
    H_neg2,
    H_neg1,
    H0,
    H1,
    H2,
    H3,
    H4]

complexities = [1/model.n_neighbors for model in models]
biases = []
variances = []
for model in models:
    decomped=[]
    decomped = bias_variance_decomp(model, X_train_knn, y_train_knn, X_test_knn, y_test, loss='mse', num_rounds=200, random_seed=123)
    biases.append(decomped[1])
    variances.append(decomped[2])
    print(decomped)
print(variances)
print(biases)
print(complexities)

plt.figure(figsize=(10, 6))

plt.plot(complexities, variances, marker='x', label='Variance')

plt.xlabel('Complexity (1 / Number of Neighbors)')
plt.ylabel(' Variance')
plt.title('Bias-Variance Tradeoff')
plt.legend()
plt.grid(True)
plt.show()
```
